```python
def prime_up_to(max_poss_prime):
    sieve=[True] * (max_poss_prime+1)
    for i in range(2,round(sqrt(max_poss_prime))+1):
        if sieve[i]:
            for j in range(2*i,len(sieve),i):
                sieve[j] = False
    return [i for i in range(2,len(sieve)) if sieve[i]]
```
