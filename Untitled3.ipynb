```python
import sys
import math

try:
    n = int(input('Enter an integer: '))
    result = ""
    sign = '+'  # 初始化sign
    D = {}      # 初始化字典
    
    if n > 0:
        # 处理正数
        temp = n
        while temp > 0:
            result = str(temp % 3) + result
            temp //= 3
        D = {0: '⇧', 1: '⬁', 2: '⬀'}
    elif n < 0:
        # 处理负数
        temp = abs(n)
        while temp > 0:
            result = str(temp % 3) + result
            temp //= 3
        sign = '-'
        D = {0: '⇩', 1: '⬂', 2: '⬃'}
    else:
        raise ValueError

    print('Your input in base 3 reads:', sign + result)
    
    # 转换数字并反转顺序
    digits = [int(c) for c in result]
    reversed_digits = digits[::-1]
    
    print('\nThis is how we will travel (reading digits from right to left):')
    print(' '.join(D[d] for d in reversed_digits))
    
    # 生成路径（统一使用反转后的顺序）
    path = ''.join(D[d] for d in reversed_digits)
    print(path)
    print('\nThis is how we travelled:')
    
    position = 0
    for direction in path:
        # 调整位置偏移
        if sign == '+':
            if direction == '⬁':   # 西北
                position -= 1
            elif direction == '⬀': # 东北
                position += 1
        else:
            if direction == '⬂':   # 东南
                position += 1
            elif direction == '⬃': # 西南
                position -= 1
        
        # 打印带缩进的方向箭头
        print(' ' * position + direction)

except ValueError:
    print('Incorrect input, giving up.')
    sys.exit()
```

```python
import sys

try:
    n = int(input('Enter an integer: '))
    if n == 0:
        raise ValueError
    
    # 初始化变量
    sign = '+' if n > 0 else '-'
    abs_n = abs(n)
    base3_digits = []
    direction_map = {
        0: ('⇧', '⇩'),
        1: ('⬁', '⬂'),
        2: ('⬀', '⬃')
    }

    # 转换为三进制（自动反转顺序）
    while abs_n > 0:
        base3_digits.append(abs_n % 3)
        abs_n //= 3
    
    # 获取方向符号（正数用索引0，负数用索引1）
    direction_symbols = [
        direction_map[d][0 if sign == '+' else 1]
        for d in base3_digits  # 已自然反转的顺序
    ]

    print(f'\nYour input in base 3 reads: {sign}{"".join(map(str, reversed(base3_digits)))}')
    
    print('\nThis is how we will travel (reading digits from right to left):')
    print(' '.join(direction_symbols))
    
    print('\nThis is how we travelled:')
    position = 0
    for symbol in direction_symbols:
        # 计算缩进位置
        print(' ' * position + symbol)
        
        # 更新位置
        if sign == '+':
            if symbol == '⬁':  # 西北
                position -= 1
            elif symbol == '⬀': # 东北
                position += 1
        else:
            if symbol == '⬂':  # 东南
                position += 1
            elif symbol == '⬃': # 西南
                position -= 1

except ValueError:
    print('Incorrect input, giving up.')
    sys.exit()
```
